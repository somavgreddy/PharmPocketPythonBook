```python
#######################################################################
# Creating tuples — use () instead of []
#######################################################################

normal_hemoglobin = (12.0, 16.0)   # (min, max) for females in g/dL
dose_schedule     = (7, 13, 21)    # Hours: morning, afternoon, night
blood_types       = ('A', 'B', 'AB', 'O')

# Accessing — same as lists
print(normal_hemoglobin[0])   # 12.0 (minimum)
print(normal_hemoglobin[1])   # 16.0 (maximum)

# Unpacking tuples — very useful!
hb_min, hb_max = normal_hemoglobin
print(f'Normal Hb range: {hb_min} – {hb_max} g/dL')

# Practical use: Check if lab value is within normal range
patient_hb = 10.5  # g/dL

if patient_hb < hb_min:
    print(f'Hemoglobin {patient_hb} g/dL is BELOW normal range — Possible anaemia')
elif patient_hb > hb_max:
    print(f'Hemoglobin {patient_hb} g/dL is ABOVE normal range')
else:
    print(f'Hemoglobin {patient_hb} g/dL is within normal range')

# Output: Hemoglobin 10.5 g/dL is BELOW normal range — Possible anaemia

# len() works on tuples too:
print(f'Number of blood types: {len(blood_types)}')  # 4

# Tuples are immutable — this would cause an ERROR:
# normal_hemoglobin[0] = 11.0   # TypeError!
```
