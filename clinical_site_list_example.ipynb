```python
# Pharmaceutical Clinical Site Deployment Python notebooks
# ClinicalSiteDeploymentNotebook.ipynb
# Initialize an active site registry list
Clinical_testing_active_sites = ["Delhi-101", "Hyd-102", "Chennai-104","Agra-200"]
print('***********************************')
print(f"Initial Active Trial Sites: { Clinical_testing_active_sites }")

# 1. Realize we missed a site and insert it in numerical order
Clinical_testing_active_sites.insert(2, "Delhi-103")
# Current list: ["Delhi-101", "Hyd-102",”Delhi-103”, "Chennai-104",”Agra-200”]
print('***********************************')
print(f" Active Trial Sites after inserting Delhi-103 : { Clinical_testing_active_sites }")

# 2. Append a newly approved regional mega-site to the end
Clinical_testing_active_sites.append("Kochi-201")

# 3. An audit reveals "Site-102" dropped out of compliance and this clinical testing site needs to be removed from the clinical site list
Clinical_testing_active_sites.remove("Hyd-102")
print('***********************************')
print(f" Active Trial Sites after removing Hyd-102 : { Clinical_testing_active_sites }")

# 4. Check where "Site-201" sits in the priority queue
priority_index = Clinical_testing_active_sites.index("Kochi-201")
print('***********************************')
print(f" Site-201 sits at index: {priority_index}")

# Print final pipeline metrics
print(f"Final Active Trial Sites: { Clinical_testing_active_sites }")
print(f"Total Active Site Count: {len(Clinical_testing_active_sites)}")  # len() is a built-in function
print(f"Site-201 Processing Index: {priority_index}")
```
