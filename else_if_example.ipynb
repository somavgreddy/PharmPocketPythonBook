```python
# if-else — does one thing if True, another if False
# Syntax:
#   if condition:
#       code when True
#   else:
#       code when False

# Example: Drug Storage Temperature Check
storage_temp = 28  # degrees Celsius
max_temp = 25      # recommended maximum

if storage_temp <= max_temp:
    print('Storage conditions: ACCEPTABLE')
    print(f'Temperature {storage_temp}°C is within the safe range.')
else:
    print('Storage conditions: UNACCEPTABLE!')
    print(f'Temperature {storage_temp}°C exceeds {max_temp}°C limit.')
    print('Action: Move drug to refrigerator immediately.')
```
