```python
# Defining a function:
# def function_name(parameter1, parameter2, ...):
#     code block
#     return value  (optional)

# Simple function called welcome_message — this takes no parameters and does not
# return anything
def welcome_message():
    print('Welcome to the Pharmacy Management System')
    print('Please consult your pharmacist for advice')

# Calling the function: Just function name and any parameters as needed.
welcome_message()   # Just use the function name with ()

# Function with parameters (inputs):
def calculate_dose(weight_kg, dose_per_kg):
#Notice how multiple lines below are commented out
    """
    Calculate total drug dose based on patient weight.
    Parameters:
        weight_kg   : Patient weight in kilograms
        dose_per_kg : Dose in mg per kg of body weight
    Returns:
        Total dose in milligrams
    """
    total_dose = weight_kg * dose_per_kg
    return total_dose

# Call the function with actual values:
dose = calculate_dose(70, 10)   # weight=70kg, dose=10mg/kg
print(f'Total dose: {dose} mg') # Output: Total dose: 700 mg

# You can call it multiple times with different values:
dose_child = calculate_dose(25, 15)
print(f'Child dose: {dose_child} mg') # Output: Child dose: 375 mg
```
