```python
# Comprehensive string Example - let's Call this notebook
# comprehensive_string_example.ipynb
#
# ============================================
# Pharmacy Prescription Label Processing Example
# Demonstrates 20+ Python string methods
# ============================================

# Raw, messy input -- exactly how data often looks when it comes to you for analysis
# from a scanned form or manual entry or excel sheets or any other means
raw_entry = "  metformin HYDROCHLORIDE 500mg  "
ndc_code = "00093-1048-01"
patient_name = "Priya Garg"
instructions = "Take 1 tablet twice daily with food"
batch_id = "487"
print("#### STEP 0: printing the raw drug entry (with trailing & beginning spaces  ) ####")
print(f"before strip(): '{raw_entry}'")

print("#### STEP 1: Let's start cleaning the raw drug entry #####")

# 1. strip() -- removes leading/trailing whitespace
cleaned = raw_entry.strip()
print(f"After strip(): '{cleaned}'")

# 2. lstrip() -- removes whitespace from the left only
print(f"lstrip(): '{raw_entry.lstrip()}'")

# 3. rstrip() -- removes whitespace from the right only
print(f"rstrip(): '{raw_entry.rstrip()}'")

# 4. lower() -- converts everything to lowercase
print(f"lower(): {cleaned.lower()}")

# 5. upper() -- converts everything to UPPERCASE (e.g. for a warning label)
print(f"upper(): {cleaned.upper()}")

# 6. title() -- capitalizes each word, good for display formatting
drug_display = cleaned.title()
print(f"title(): {drug_display}")

# 7. swapcase() -- flips upper/lower case, just to demonstrate
print(f"swapcase(): {cleaned.swapcase()}")

print("\n----****** STEP 2: let's now start checking and searching the drug name ----******")

# 8. startswith() -- checks if the string starts with something
print(f"startswith('metformin'): {cleaned.lower().startswith('metformin')}")

# 9. endswith() -- checks if the string ends with something
print(f"endswith('500mg'): {cleaned.lower().endswith('500mg')}")

# 10. find() -- returns the position of a substring (-1 if not found)
position = cleaned.lower().find("hydrochloride")
print(f"find('hydrochloride'): position {position}")

# 11. rfind() -- finds the position of the LAST occurrence
print(f"rfind('m'): position {cleaned.lower().rfind('m')}")

# 12. count() -- counts how many times something appears
print(f"count('m'): {cleaned.lower().count('m')}")

print("\n#### STEP 3: Splitting and rebuilding text ####")

# 13. split() -- breaks the string into a list of words
words = cleaned.split()
print(f"split(): {words}")

# 14. join() -- glues a list back together with a separator
rejoined = "-".join(words)
print(f"join(): {rejoined}")

# 15. rsplit() -- splits from the right, useful for limited splits
ndc_parts = ndc_code.rsplit("-", 1)
print(f"rsplit('-', 1) on NDC: {ndc_parts}")

# 16. partition() -- splits into exactly 3 parts: before, separator, after
before, sep, after = ndc_code.partition("-")
print(f"partition('-'): before='{before}', sep='{sep}', after='{after}'")

print("\n#### STEP 4: Validating characters (useful for form checks) ####")

# 17. isalpha() -- True if ALL characters are letters
print(f"isalpha() on 'metformin': {'metformin'.isalpha()}")

# 18. isdigit() -- True if ALL characters are digits
print(f"isdigit() on '500': {'500'.isdigit()}")

# 19. isalnum() -- True if characters are letters AND/OR digits (no spaces/symbols)
print(f"isalnum() on 'batch487': {'batch487'.isalnum()}")

# 20. isspace() -- True if the string is ONLY whitespace
print(f"isspace() on '   ': {'   '.isspace()}")

print("\n#### STEP 5: Formatting the final printed label ####")

# 21. replace() -- swaps one substring for another
short_name = drug_display.replace("Hydrochloride", "HCl")
print(f"replace(): {short_name}")

# 22. zfill() -- pads a number with leading zeros (e.g. for batch IDs)
padded_batch = batch_id.zfill(6)
print(f"zfill(6): {padded_batch}")

# 23. center() -- centers text within a given width, useful for printed labels
print(f"center(40): '{short_name.center(40, '-')}'")

# 24. format() -- inserts values into a template string
label_template = "{name} | {drug} | Batch: {batch}"
final_label = label_template.format(
    name=patient_name.title(),
    drug=short_name,
    batch=padded_batch
)
print(f"format(): {final_label}")

print("\n**** FINAL PRESCRIPTION LABEL TO BE PRINTED ON THE INVOICE ****")
print(final_label)
```
