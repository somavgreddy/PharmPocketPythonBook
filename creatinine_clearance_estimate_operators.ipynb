```python
#######################################################
# Arithmetic operators for pharmaceutical calculations
#######################################################

# Basic operations
addition       = 500 + 250     # Result: 750 mg (total dose)
subtraction    = 1000 - 250    # Result: 750 mg (remaining)
multiplication = 10 * 70      # Result: 700 mg (dose × weight)
division       = 700 / 2      # Result: 350.0 mg (split dose)  — always gives float
floor_division = 700 // 2     # Result: 350    (integer division, no decimal)
modulus        = 700 % 3      # Result: 1 (remainder after dividing 700 by 3)
exponent       = 2 ** 3       # Result: 8 (2 to the power 3)

print('Daily dose:', multiplication, 'mg')
print('Per dose:', division, 'mg')

# Practical Example: Creatinine Clearance Estimate
# Using simplified Cockcroft-Gault formula
age = 55
weight = 70      # kg
serum_creatinine = 1.2   # mg/dL
patient_sex="F"
#formula for CrCLis given below

# CrCl=72×Serum Creatinine(mg/dL)(140−Age)×Weight(kg)×(0.85 if female)​

# CrCl = ((140 - age) * weight) / (72 * serum_creatinine)
crcl = ((140 - age) * weight) / (72 * serum_creatinine)
#Now adjust the calculated crcl for females as the patient is female in our case
if patient_sex == "F":
    crcl = crcl * 0.85
print(f'Creatinine Clearance: {crcl:.1f} mL/min')
# Output: Creatinine Clearance: 45.1 mL/min
```
