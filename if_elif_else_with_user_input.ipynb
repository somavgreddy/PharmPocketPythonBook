```python
# if-elif-else — handles multiple conditions
# Syntax:
#   if condition1:
#       code block 1
#   elif condition2:
#       code block 2
#   elif condition3:
#       code block 3
#   else:
#       default code block

# Example: Paracetamol Dose Recommendation Based on Age Group
#The line below will open a user inout dialogue so that user can enter patient age dynamically at run time
#pay attention to the casting being done once the user input is entered. if you enter non-numeric values, this below line will throw an error
age = int(input('Enter patient age: '))

if age < 1:
    print('Consult paediatrician — not recommended under 1 year')
elif age >= 1 and age <= 5:
    dose = 120  # mg
    print(f'Recommended Paracetamol dose: {dose} mg')
elif age >= 6 and age <= 12:
    dose = 250  # mg
    print(f'Recommended Paracetamol dose: {dose} mg')
elif age >= 13 and age <= 17:
    dose = 500  # mg
    print(f'Recommended Paracetamol dose: {dose} mg')
else:
    dose = 500  # to 1000 mg
    print(f'Recommended Paracetamol dose: {dose} mg to 1000 mg')
```
