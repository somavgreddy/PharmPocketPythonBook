```python
#####################################################################
# Default parameters — used when the caller does not provide a value
#####################################################################

# Syntax: def function(param=default_value)

def calculate_creatinine_clearance(age, weight, serum_cr, gender='male'):
    """
    Cockcroft-Gault formula for Creatinine Clearance
    Parameters:
        age       : Patient age in years (required)
        weight    : Body weight in kg (required)
        serum_cr  : Serum creatinine in mg/dL (required)
        gender    : 'male' or 'female' (optional, default='male')
    Returns:
        Creatinine clearance in mL/min
    """
    crcl = ((140 - age) * weight) / (72 * serum_cr)

    if gender == 'female':
        crcl *= 0.85   # Multiply by 0.85 for female patients

    return round(crcl, 1)

# Call without gender (uses default 'male'):
male_crcl = calculate_creatinine_clearance(55, 70, 1.2)
print(f'Male CrCl: {male_crcl} mL/min')   # Male CrCl: 45.1 mL/min

# Call with gender specified:
female_crcl = calculate_creatinine_clearance(55, 70, 1.2, gender='female')
print(f'Female CrCl: {female_crcl} mL/min') # Female CrCl: 38.3 mL/min

# Using keyword arguments — order does not matter:
result = calculate_creatinine_clearance(weight=65, age=60, serum_cr=1.5, gender='female')
print(f'CrCl: {result} mL/min')
```
