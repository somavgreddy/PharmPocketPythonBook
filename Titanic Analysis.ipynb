```python
max_age_survived = titanic_df[titanic_df['survived'] == 1]['age'].max()
print(f"The maximum age of a person who survived is: {max_age_survived}")
```

```python
min_age_survived = titanic_df[titanic_df['survived'] == 1]['age'].min()
print(f"The min age of a person who survived is: {min_age_survived*12}")


#Now adding more condition like max & min of females who survived
max_age_female_survived = titanic_df[(titanic_df['survived'] == 1) & (titanic_df['sex'] == 'female')]['age'].max()
print(f"The maximum age of a female who survived is: {max_age_female_survived}")

min_age_female_survived = titanic_df[(titanic_df['survived'] == 1) & (titanic_df['sex'] == 'female')]['age'].min()
print(f"The min age (in months) of a female who survived is: {min_age_female_survived*12}")

females_survived = titanic_df[(titanic_df['survived'] == 1) & (titanic_df['sex'] == 'female')]['sex'].count()
print(f"The number of females who survived is: {females_survived}")

#Now compute the % of males who survived & % females who survived

males_survived = titanic_df[(titanic_df['survived'] == 1) & (titanic_df['sex'] == 'male')]['sex'].count()
print(f"The number of males who survived is: {males_survived}")


max_age_male_survived = titanic_df[(titanic_df['survived'] == 1) & (titanic_df['sex'] == 'male')]['age'].max()
print(f"The maximum age of a male who survived is: {max_age_female_survived}")

min_age_male_survived = titanic_df[(titanic_df['survived'] == 1) & (titanic_df['sex'] == 'male')]['age'].min()
print(f"The min age (in months) of a male who survived is: {min_age_female_survived*12}")

```

```python
# Calculate total number of females and males
total_females = titanic_df[titanic_df['sex'] == 'female']['sex'].count()
print(f"Total number of females: {total_females}")
total_males = titanic_df[titanic_df['sex'] == 'male']['sex'].count()
print(f"Total number of males: {total_males}")

total_passengers = titanic_df['sex'].count()
print(f"Total number of passengers: {total_passengers}")

# Calculate percentage of females who survived
percent_females_survived = (females_survived / total_females) * 100
print(f"Percentage of females who survived: {percent_females_survived:.2f}%")

# Calculate percentage of males who survived
percent_males_survived = (males_survived / total_males) * 100
print(f"Percentage of males who survived: {percent_males_survived:.2f}%")
```

```python
#now calculate total number of children below the age of 10 yrs
total_children = titanic_df[titanic_df['age'] < 10]['age'].count()
print(f"Total number of children below the age of 10 years: {total_children}")
#Now how many of them survived
children_survived = titanic_df[(titanic_df['age'] < 10) & (titanic_df['survived'] == 1)]['age'].count()
print(f"Number of children below the age of 10 years who survived: {children_survived}")
```
