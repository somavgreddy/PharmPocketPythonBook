```python
########################################################################
# user enters an input number & then the program finds out the value
# of the factorial
# the same code to handle exceptions if user enters non-numeric values
########################################################################
try:
  user_input_str = input("Enter a number to calculate its factorial: ")
  user_input = int(user_input_str)
  print(f"Let us now try to build the factorial value of: {user_input}")
  # Logic to create factorial
  # Define a function in python called factorial to calculate the factorial value of the entered number
  def factorial(num):
    if num < 0:
        return "Factorial is not defined for negative numbers"
    elif num == 0:
        return 1
    else:
        result = 1
        for x in range(1, num + 1):
            result *= x
        return result
  #Function definition is complete & let's now invoke the method
  fact = factorial(user_input)
  print("########################################################################")
  print(f"Factorial of the number {user_input} is {fact}")
except ValueError:
  print("Invalid input. Please enter a numeric value.")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```

```python
########################################################################
# user enters an input number & then the program Builds a right angle traingle with as many *
# the same code to handle exceptions if user enters non-numeric values
########################################################################
try:
  user_input_str = input("Enter a number to draw a right-angle triangle: ")
  user_input = int(user_input_str)
  print(f"Let us now try to build the triangle with : {user_input}")
  # Logic to create a right-angle triangle
  STAR='*'
  def print_triangle(num):
    if num <= 0:
        print("Not a valid number. Can't draw a triangle with non-positive numbers.")
    else:
        for i in range(1, num + 1):
            print(STAR * i)
  #This function does not return any value back - it just prints
  #Function definition is complete & let's now invoke the method
  print_triangle(user_input)

except ValueError:
  print("Invalid input. Please enter a numeric value.")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```

# Try -Exception explanation

The try-except block in Python is used for error handling. Code that might raise an exception is placed inside the try block. If an exception occurs, the code inside the corresponding except block is executed, allowing the program to handle the error gracefully instead of crashing.

In the below code , think what happens when you enter a non -numeric input?

With out Try -Exception blocks , The code will abruptly fail.

```python
########################################################################
# user enters an input number & then the program Builds Fibonacci series until the given number
# the same code to handle exceptions if user enters non-numeric values
########################################################################
try:
  user_input_str = input("Enter a number to generate Fibonacci series up to that number: ")
  user_input = int(user_input_str)
  print(f"Let us now build the Fibonacci series up to: {user_input}")

  # Define a function to create fibonacci series now
  def fibonacci_series(num):
    if num <= 0:
      print("Please enter a positive integer.")
      return
    elif num == 1:
      print("0")
      return
    else:
      a, b = 0, 1 # This is claeed simultaneous assignment in Python. This means a=0 and b=1
      series = [a] # This is a list called series with initial value as "0"
      while b <= num:
        series.append(b)
        a, b = b, a + b
        # This above statement is called simultaneous assignment in Python
        # evaluate the right side first
        # So first it takes current value of a & b and computes a+b (0+1=1)
        # after executing this line a=1,b=1

        # The above line technically means the below 2 lines combined in to one line
        #a=b
        #b=a+b

      print("Fibonacci series:")
      print(series)
  # End of function definition - Let us now invoke the function

  fibonacci_series(user_input)
  # when a Python list is printed using print(), the default string representation of the list separates the values with commas.

except ValueError:
  print("Invalid input. Please enter a numeric value.")
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```
