```python
###############################################################################
# while loop — repeats as long as condition is True
# let’s call it while_loop_example.ipynb
# Syntax:
#   while condition:
#       code to repeat
#       (must update condition inside loop to avoid infinite loop!)

# Example: Drug Dispensing System
# Keep asking user's dynamic input for prescriptions until pharmacist types 'quit'
# quit acts as the breaker of the while loop
###############################################################################
prescription_count = 0
print('Pharmacy Dispensing System')
print('Type drug name or type quit to exit')
while True:
    drug = input('Enter drug name: ') # Dynamic user input will default to sting in Python
    # if you need the dynamic user input to be a number (say integer or float etc), Then you would need to cast the input value
    # remember we are using string's lower() method to convert user input string in to all lower characters
    if drug.lower() == 'quit':
        break   # Exit the loop
    prescription_count += 1
    print(f'Dispensed: {drug} (Prescription #{prescription_count})')
print(f'\nTotal prescriptions dispensed today: {prescription_count}')
```

```python
# Example: Simulate Drug Concentration Over Time over 24 hr period
# As time increases --> The drug concentration decreases and in this example , We use the concept of
# Half life hours
concentration = 100.0  # mg/L initial concentration
half_life_hours = 4    # hours
time_hours = 0
print('Time (hr) | Concentration (mg/L)')
while concentration >= 1.0:  # Continue until concentration < 1 mg/L
    print(f'{time_hours:8} | {concentration:.2f}')
    concentration /= 2       # Each half-life reduces by half
    time_hours += half_life_hours
```

```python
import datetime

# break — immediately exits the loop
# continue — skips the rest of current iteration and goes to next
# This is a for loop example
# Example: break — find first expired drug
drugs = [
    {'name': 'Amoxicillin', 'expiry': 2026},
    {'name': 'Metformin',   'expiry': 2024},   # Expired!
    {'name': 'Aspirin',     'expiry': 2027},
    {'name': 'Ibuprofen',   'expiry': 2028},
    {'name': 'Acetaminophen','expiry': 2029},
    {'name': 'Azithromycyn','expiry': 2025}    #Expired
]
#print(drugs)
current_year = datetime.datetime.now().year

for drug in drugs:
    if drug['expiry'] < current_year:
        print(f"EXPIRED: {drug['name']} expired in {drug['expiry']}")
        break  # This stops checking after first expired drug
        #if break is commented , Then the loop will not break at the first EXPIRED drug
        #it will iterate over the entire drug list and pick Axithromycyn expiry as well
    print(f"{drug['name']} is valid until {drug['expiry']}")

# Output:
# Amoxicillin is valid until 2026
# EXPIRED: Metformin expired in 2024
# Aspirin is valid until 2026
# Ibuprofen is valid until 2028
# Acetaminophen is valid until 2029
# EXPIRED: Azithromycyn expired in 2025

# Example: continue — skip drugs with missing data
drug_doses = [500, None, 250, None, 750,1000,500]

for dose in drug_doses:
    if dose is None:
        print('Skipping record with missing dose')
        continue    # Go to next iteration
    adjusted_dose = dose * 0.8   # 80% dose for elderly
    print(f'Adjusted dose: {adjusted_dose} mg')
```
