```python
################################################################
#A hospital network has two pharmacies that each maintain their own list of patient-reported
#allergies for a shared patient (since the patient has visited both locations).
#Write a program that takes both allergy lists and identifies:
#(a) allergies reported at both pharmacies (confirmed, high-confidence allergies), and
#(b) allergies reported at only one location (which may need follow-up verification
# before the patient's next prescription is filled).
################################################################
#let's define a function called check_allergy_overlap that takes 2 parameters
def check_allergy_overlap(pharmacy_a, pharmacy_b):
    """
    Compares allergy records from two pharmacies for a shared patient.

    Parameters:
        pharmacy_a (set): allergies reported at Pharmacy A
        pharmacy_b (set): allergies reported at Pharmacy B

    Returns:
        dict: confirmed (both), needs_verification (only one), master_list (all, deduplicated)
    """
    confirmed = pharmacy_a & pharmacy_b          # intersection: reported at BOTH locations
    needs_verification = pharmacy_a ^ pharmacy_b  # symmetric difference: reported at only ONE location
    master_list = pharmacy_a | pharmacy_b         # union: complete deduplicated allergy list

    return {
        "confirmed": confirmed,
        "needs_verification": needs_verification,
        "master_list": master_list,
    }


# Example data — allergy records for the same patient at two pharmacy locations
pharmacy_a = {"Penicillin", "Sulfa", "Latex"}
pharmacy_b = {"Penicillin", "Aspirin"}

result = check_allergy_overlap(pharmacy_a, pharmacy_b)

print("Pharmacy A reported allergies:", pharmacy_a)
print("Pharmacy B reported allergies:", pharmacy_b)
print()
print("✅ Confirmed allergies (reported at both pharmacies):")
print(" ", result["confirmed"])

print()
print("⚠️  Needs verification (Allergies reported at only one pharmacy):")
print(" ", result["needs_verification"])

print()
print("📋 Unified master allergy list (for patient's combined record):")
print(" ", result["master_list"])
```
