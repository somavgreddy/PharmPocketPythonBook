```python
# Clinical Trial Example that involves most of the data types of Python-let’s call this program “datatypes_clinical_trial_patient_data.ipny”
# This example uses most of the Python data types

import datetime

# ==============================================================================
# 1. INITIALIZING DATA TYPES Clinical Trial Context)
# ==============================================================================

# A. Numeric Types
patient_age: int = 45                # int: Whole number for demographic tracking
dosage_mg: float = 250.50            # float: Decimal value for exact active compound metric
cellular_resonance: complex = 3 + 4j # complex: Used in advanced medical imaging/signal processing

# B. Sequence Types
protocol_id: str = "NVS-2026-ONC"    # str: Text identifier for the clinical trial protocol
administered_meds: list = ["Therapy-A", "Placebo"]     # list: Mutable sequence tracking daily intervention arm elements
patient_geo_location: tuple = (40.5188, -74.3322)      # tuple: Immutable coordinates (Latitude, Longitude) of trial site

# C. Mapping Type
patient_vitals: dict = {                               # dict: Key-value mapping for a comprehensive electronic health record
    "patient_id": "PAT-101",
    "blood_pressure": "120/80",
    "heart_rate": 72,
    "is_responsive": True
}

# D. Set Types
observed_side_effects: set = {"Nausea", "Fatigue"} # set: Unordered collection of unique, non-duplicate clinical symptoms
# Adding a duplicate to demonstrate set properties (will be automatically deduplicated)
observed_side_effects.add("Fatigue")

locked_genetic_markers: frozenset = frozenset(
    ["BRCA1_Mut", "ALK_Pos"]
) # frozenset: Immutable set representing fixed baseline genome exclusions

# E. Boolean Type
is_gxp_compliant: bool = True                          # bool: Logical state verifying strict FDA/EMA data compliance


# F. Date Object (Standard Library Object Add-on)
trial_start_date: datetime.date = datetime.date(2026, 6, 7) # datetime.date object: Standard date tracking wrapper

# G. Binary Types (Added to cover all common types)
raw_protocol_signature: bytes = b'\x45\x23\x78\x11' # bytes: Immutable sequence of bytes for cryptographic signatures
mutable_telemetry_buffer: bytearray = bytearray(b'Initial Data') # bytearray: Mutable sequence of bytes for sensor stream data


# ==============================================================================
# 2. RUNNING PROCESSING LOGIC & LOGGING RESULTS
# ==============================================================================

print("" * 50)
print(f"CLINICAL RUN RECORD: {protocol_id} | DATE: {trial_start_date}")
print("" * 50)

# Outputting values alongside verification of their native Python Data Types
print(f"[NUMERIC] Patient Age: {patient_age} | Type: {type(patient_age)}")
print(f"[NUMERIC] Compound Dosage: {dosage_mg} mg | Type: {type(dosage_mg)}")
print(f"[NUMERIC] Biosensor Signal Phase: {cellular_resonance} | Type: {type(cellular_resonance)}")
print()
print(f"[SEQUENCE] Protocol Identifier: {protocol_id} | Type: {type(protocol_id)}")
print(f"[SEQUENCE] Active Treatment Queue (Mutable): {administered_meds} | Type: {type(administered_meds)}")
print(f"[SEQUENCE] Trial Site Coordinates (Immutable): {patient_geo_location} | Type: {type(patient_geo_location)}")
print()
print(f"[MAPPING] Patient Core Health Metrics Profile:\n   {patient_vitals}\n   Type: {type(patient_vitals)}")
print()
print(f"[SET] Active Adverse Events Encountered (Unique): {observed_side_effects} | Type: {type(observed_side_effects)}")
print(f"[SET] Baseline Genomic Exclusions (Locked): {set(locked_genetic_markers)} | Type: {type(locked_genetic_markers)}")
print()
print(f"[BOOLEAN] System Regulatory GxP Integrity Validated: {is_gxp_compliant} | Type: {type(is_gxp_compliant)}")
print()
print(f"[BINARY] Crypto-Validation Signature (Raw Bytes): {raw_protocol_signature} | Type: {type(raw_protocol_signature)}")
print(f"[BINARY] Telemetry Stream State (Bytearray Pre-Update): {mutable_telemetry_buffer.decode()} | Type: {type(mutable_telemetry_buffer)}")

print("-" * 80)
print("PROCESSING COMPLETE: All health records successfully structured, compiled, and mapped.")
print("-" * 80)
```
