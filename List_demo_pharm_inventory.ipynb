```python
# ============================================
# Pharmacy Inventory Management System
# Demonstrates all 11 Python list methods
# Let's call it List_demo_pharm_inventory.ipynb
# ============================================

# Starting inventory of drugs in stock
inventory = ["Aspirin", "Ibuprofen", "Metformin","Tyleonol","Mucinex","Zirtec","Azithromycin"]
print(f"Starting inventory: {inventory}")

print("\n---- Receiving new stock ----")

# 1. append() -- add a single new drug to the end of the list
inventory.append("Lisinopril")
print(f"After append('Lisinopril'): {inventory}")

# 2. extend() -- add multiple new drugs at once (e.g. a bulk delivery)
inventory.extend(["Atorvastatin", "Omeprazole"])
print(f"After extend([...]): {inventory}")

# 3. insert() -- add a drug at a specific position (e.g. priority placement)
inventory.insert(0, "Epinephrine")   # emergency drug placed first
print(f"After insert(0, 'Epinephrine'): {inventory}")

print("\n---- Checking stock details ----")

# 4. index() -- find the position of a specific drug
position = inventory.index("Metformin")
print(f"index('Metformin'): found at position {position}")

# 5. count() -- check how many times a drug appears (duplicate entry check)
inventory.append("Aspirin")   # simulate a duplicate entry by mistake
duplicate_count = inventory.count("Aspirin")
print(f"count('Aspirin'): appears {duplicate_count} times")

print("\n---- Dispensing and removing stock ----")

# 6. remove() -- remove a specific drug by name (e.g. recalled medication)
inventory.remove("Aspirin")   # removes only the FIRST occurrence
print(f"After remove('Aspirin') [removes first match]: {inventory}")

# 7. pop() -- remove and return an item (e.g. dispensing the last-added drug)
dispensed_drug = inventory.pop()
print(f"After pop() -> dispensed: '{dispensed_drug}', remaining: {inventory}")

# pop() with an index -- dispense a drug from a specific position
dispensed_specific = inventory.pop(1)
print(f"After pop(1) -> dispensed: '{dispensed_specific}', remaining: {inventory}")

print("\n---- Organizing the shelf ----")

# 8. sort() -- arrange drugs alphabetically for easy shelf browsing
inventory.sort()
print(f"After sort(): {inventory}")

# 9. reverse() -- reverse the current order (e.g. show newest-sorted-last first)
inventory.reverse()
print(f"After reverse(): {inventory}")

print("\n---- Backup and audit ----")

# 10. copy() -- create a backup snapshot of current inventory before an audit
inventory_backup = inventory.copy()
print(f"Backup created (copy()): {inventory_backup}")

# Prove it's an independent copy -- modifying original doesn't affect backup
inventory.append("Ranitidine")
print(f"Live inventory after new addition: {inventory}")
print(f"Backup remains unchanged: {inventory_backup}")

print("\n---- End of day: clearing a recalled batch list ----")

recalled_batch_list = ["BATCH-2026-0417", "BATCH-2026-0418"]
print(f"Recalled batch list before clear(): {recalled_batch_list}")

# 11. clear() -- empty out the list completely (e.g. recall list processed and closed)
recalled_batch_list.clear()
print(f"After clear(): {recalled_batch_list}")

print("\n---- FINAL LIVE INVENTORY ----")
print(inventory)
```
