```python
##################################################################
# Creating a dictionary:
##################################################################

patient = {
    'patient_id': 'MH-2024-001',
    'name': 'Ananya Krishnan',
    'age': 42,
    'weight_kg': 58.5,
    'diagnosis': 'Type 2 Diabetes',
    'medications': ['Metformin 500mg', 'Glipizide 5mg'],
    'allergies': ['Sulfa drugs'],
    'is_pregnant': False
}

# Accessing values using keys:
print(patient['name'])         # Ananya Krishnan
print(patient['age'])          # 42
print(patient['medications'])  # ['Metformin 500mg', 'Glipizide 5mg']

# Safe access using .get() — returns None if key doesn't exist
print(patient.get('blood_group', 'Not recorded'))   # 'Not recorded'

# Adding or updating entries:
patient['blood_group'] = 'B+'       # Add new key
patient['weight_kg']   = 59.0       # Update existing key

# Remove an entry:
patient.pop('is_pregnant')

# Looping through a dictionary:
for key, value in patient.items():
    print(f'  {key}: {value}')

# Check if a key exists:
if 'allergies' in patient:
    print('ALLERGY ALERT:', patient['allergies'])

# Dictionary of drug prices (formulary lookup):
drug_prices = {
    'Metformin 500mg': 12.50,
    'Aspirin 75mg': 5.00,
    'Atorvastatin 10mg': 45.00,
}

drug = 'Aspirin 75mg'
print(f'{drug} costs Rs. {drug_prices.get(drug, "Not in formulary")}')
```
