```python
import csv

# Data to write — a list of patient records
patients = [
    ['PatientID', 'Name', 'Age', 'Drug', 'Dose_mg', 'Duration_days'],
    ['P001', 'Rohit Sharma', 45, 'Metformin', 500, 90],
    ['P002', 'Sunita Patel', 38, 'Atorvastatin', 10, 180],
    ['P003', 'Mohan Das', 62, 'Amlodipine', 5, 365],
    ['P004', 'Lakshmi Iyer', 55, 'Aspirin', 75, 180],
]

# Writing to CSV file:
# open(filename, mode, newline, encoding)
# mode: 'w' = write (creates new or overwrites), 'a' = append
with open('patient_records.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerows(patients)    # Write all rows at once

print('File written successfully!')
```
