```python
#######################################################
# Type Casting Example — Drug Dosage Calculator
#######################################################
# Suppose the user enters data as text (string):
weight_str = '70'         # This is a string, not a number yet!
dose_per_kg_str = '10'   # Also a string

# Convert strings to numbers for calculation:
weight_kg = float(weight_str)     # str → float: '70' becomes 70.0
dose_per_kg = int(dose_per_kg_str) # str → int: '10' becomes 10

# Now we can calculate:
total_dose_mg = weight_kg * dose_per_kg
print('Total dose:', total_dose_mg, 'mg')   # Output: Total dose: 700.0 mg

# Other type conversions:
age = int(45.7)          # float → int: becomes 45 (decimal part removed)
price = str(299.99)      # float → str: becomes '299.99'
is_valid = bool(1)       # int → bool: 1 becomes True, 0 becomes False

# Check the type after conversion:
print(type(total_dose_mg))   # <class 'float'>
print(type(price))           # <class 'str'>
print(type(is_valid))        # <class 'bool
```
